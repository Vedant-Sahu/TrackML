```python
import pandas as pd
import tensorflow as tf
import numpy as np

#plotting 
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
%matplotlib notebook
```

```python
df = pd.read_csv('train_100_events/event000001099-particles.csv')
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(df.vx,df.vy,df.vz, 'o')
```

```python
ax.plot_wireframe(df.vx, df.vy, df.vz, rstride=5, cstride=5)
# rotate the axes and update
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```
